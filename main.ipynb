5- Geographical Location:

In order to figure out the prices of the houses with the variables we have, we are required to specify a fixed area on the map since prices could vary largely depending on the location of the house

We will add the distance function given 2 latitudes and longitudes 

```python
from math import sin, cos, sqrt, atan2
def distance(lat1,lon1,lat2,lon2):
    R = 6373.0
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = (sin(dlat/2))**2 + cos(lat1) * cos(lat2) * (sin(dlon/2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance
```

We will have a on the location where all the houses are located by finding the min and max longitudes and latitudes:

```python
df1[['long','lat']].agg(['min','max'])
```

6- Estimation of prices 

In the final step, we will consider that we have a missing price for a certain house and we want to estimate the fair price based on our previous analysis. 
We will take a random price value from the df_area2 dataFrame and we will insert a missing value instead

```python
import numpy as np
missing_percentage = 0.1  # 10% missing values
```

```python
df_missing = df_area1.copy()
```

```python
mask = np.random.rand(*df_missing['price'].shape)<missing_percentage
df_missing['price'][mask] = np.nan
```

```python
df_missing
```

So now we have a missing price on the ID no. 7853230200 with a sqft_living=2480.
Following our study, we have to fill in the price as equal to sqft_living * average price/sqft: 2480 * 193

```python
avg_price_sqft=193 # The average price per sqft
```

```python
df_filled=df_missing.copy()
```

```python
df_filled['price'] = df_missing['price'].fillna(df_missing['sqft_living']*avg_price_sqft)
```

```python
df_filled
```

As shown in the table, the filled value is 478640 which is an estimation of the price based on the living area.
The initial price before inserting the missing value was 470000 which gives an acceptable estimation with 2% margin of error.

#### Price binning

In order to simplify our analysis it is better to distribute the price into different categories: A,B,C,D from the most to the least expensive

```python
def condition(x):
    if (x>=75000 and x<200000): 
        return 'E' 
    elif (x>=200000 and x<400000): 
        return 'D' 
    elif (x>=400000 and x<800000): 
        return 'C'
    elif (x>=800000 and x<1200000):
        return 'B'
    else:
         return 'A'
```

```python
df2=df.copy()
```

```python
df2['Price_Category']=df2['price'].apply(condition)
```

```python
df2['Price_Category'].value_counts(normalize=True).mul(100)
```

```python
df2[['price','bedrooms','sqft_living','grade','condition','sqft_basement','Price_Category']].groupby('Price_Category').agg({ 'sqft_living' :'mean','grade':'median','condition':'median','bedrooms':'median','sqft_basement':'mean'}).reset_index()
```

This table shows that both the living area and the grade are directly related and have the biggest impact on the prices, along with the location.However regarding the condition, it does no interfere as much regarding the price catogeries but could still have a small impact on the houses where the prices are relatively close
